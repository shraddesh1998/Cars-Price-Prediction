## Merging Dataset: (Append)

```python
new_delhi = pd.read_csv('new_delhi.csv')
mumbai = pd.read_csv('mumbai.csv')
kochi = pd.read_csv('kochi.csv')
jaipur = pd.read_csv('jaipur.csv')
kolkata = pd.read_csv('kolkata.csv')
lucknow = pd.read_csv('lucknow.csv')
chennai = pd.read_csv('chennai.csv')
hydrebad = pd.read_csv('hydrabad.csv')
bangalore = pd.read_csv('bangalore.csv')
indore = pd.read_csv('indore.csv')
pune = pd.read_csv('pune.csv')
ahmedabad = pd.read_csv('ahmedabad.csv')
```

```python
new_delhi
```

```python
c1 = new_delhi.append(mumbai, ignore_index = True)
c2 = c1.append(kochi, ignore_index = True)
c3 = c2.append(jaipur, ignore_index = True)
c4 = c3.append(kolkata, ignore_index = True)
c5 = c4.append(lucknow, ignore_index = True)
c6 = c5.append(chennai, ignore_index = True)
c7 = c6.append(hydrebad, ignore_index = True)
c8 = c7.append(bangalore, ignore_index = True)
c9 = c8.append(indore, ignore_index = True)
c10 = c9.append(pune, ignore_index = True)
cars24 = c10.append(ahmedabad, ignore_index = True)
```

```python
cars24
```

```python
cars = cars24.sample(frac = 1).reset_index()
cars2 = cars.drop(['index'], axis = 1)
cars2
```

```python
cars2.to_csv('scraped cars24 data (1).csv')
```
